```python
#Import libraries
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
img=cv2.imread('images/caves.jpg')
```

```python
plt.imshow(img)
```

```python
img.shape
```

# Slicing the image

```python
new_img=img[100:300,100:300]
```

```python
plt.imshow(new_img)
```

# To covert image from BGR to RGB

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

# Relative cropping

```python
height,width, _ =img.shape
```

```python
height
```

```python
width
```

```python
img_new=img[80:454,60:320]
img_new=cv2.cvtColor(img_new,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_new)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.title('Original Image'+str(rgb_img.shape))
plt.imshow(rgb_img)
plt.subplot(1,2,2)
plt.title('Cropped Image'+str(img_new.shape))
plt.imshow(img_new)
```
