## Growth Curves

### Imports

```python
### -------------------
### Importing libraries
### -------------------

import numpy as np

import matplotlib.pyplot as plt
```

### Global variables

### Input data

```python
### ----------
### input data
### ----------

x = np.arange(-5 , 5 , 0.25)
```

```python
x
```

## 1. Linear Growth

```python
# Linear equation

y = x * 3.47 + 0.56
```

```python
plt.plot(y);
```

## 2. Exponential Growth / Polynomial Growth

```python
# Polynomial Growth

y = x ** 2 + 0.89

plt.plot(y)
```

```python
y 
```

##  3. Logarithmic growth

```python
x = np.arange(1, 10, 0.25)
```

```python
# Logarithmic growth

y = np.log(x)

plt.plot(y);
```

## 4 Sigmoid Growth

```python
# defining sigmoid function

def fun_sigmoid(x):
    
    return 1 / (1 + np.exp(-x))
```

```python
# Sigmoid Growth

x = np.arange(-10, 10, 0.25)

y = fun_sigmoid(x)

plt.plot(y);
```

## 5. Tanh Growth

```python
def fun_tanh(x):
    
    return np.tanh(x)
```

```python
# tanh growth

x = np.arange(-10, 10, 0.25)

y = fun_tanh(x)

plt.plot(y);
```
