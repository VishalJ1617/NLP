# prepare the data

```python
alphabets='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabets)
```

# Preprocessing the data

```python
int_to_char=dict(enumerate(alphabets))
```

```python
int_to_char
```

```python
char_to_int=dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int
```

```python
int_to_char[5]
```

```python
seq_lenght=1
```

```python
x=[]
y=[]
for i in range(len(alphabets)-seq_lenght):
    seq_in=[alphabets[i]]
    seq_out=alphabets[i+1]
    x+=[[char_to_int[seq_in[0]]]]
    y+=[char_to_int[seq_out]]
    print(seq_in,'-->',seq_out)
```

# One hot encoding

```python
from keras.utils import to_categorical
```

```python
y_new=to_categorical(y)
```

```python
y_new
```

```python
y_new.shape    
```

```python
import numpy as np
```

```python
np.array(x).shape
```

```python
x=np.reshape(x,[25,1,1])
```

```python
x.shape
```

```python
x
```

```python
x=x/25
```

```python
x
```

# Build the model

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
import numpy as np
```

```python
model=Sequential()
model.add(Dense(25,input_shape=(1,1),activation='relu'))
model.add(LSTM(128))
model.add(Dense(26,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```

# Seq lenght=3

```python
seq_lenght=3
```

```python
x=[]
y=[]
for i in range(len(alphabets)-seq_lenght):
    seq_in=alphabets[i:i+seq_lenght]
    seq_out=alphabets[i+seq_lenght]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in,'-->',seq_out)
```

```python
x
```

```python
y_new=to_categorical(y)
y_new
```

```python
np.array(x).shape
```

```python
x=np.reshape(x,(len(x),seq_lenght,1))
x=x/25
x.shape
```

```python
model=Sequential()
model.add(Dense(23,input_shape=(3,1),activation='relu'))
model.add(LSTM(128))
model.add(Dense(26,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```

```python
def prediction(news):
    val=np.array([[char_to_int[x]] for x in news])
    val=np.reshape(val,(1,3,1))
    val=val/25
    result=model.predict(val,verbose=False)
    result=int_to_char[np.argmax(result)]
    return result
```

```python
prediction('VWX')
```
