##  Symantic analysis

```python
### -------------------
### Importing libraries
### -------------------

from nltk.corpus import wordnet

import numpy as np
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
# .synset generate list of all  the meaning of the word provided as parameter

x = wordnet.synsets('create')

x
```

```python
synonyms_create = set()

for word in wordnet.synsets('create'):
    
    synonyms_create.add(word.lemmas()[0].name())
    
synonyms_create
```

```python
synonyms_prompt = set()

for word in wordnet.synsets('prompt'):
    
    synonyms_prompt.add(word.lemmas()[0].name())
    
synonyms_prompt
```

```python
synonyms_prompt = set()

for word in wordnet.synsets('sky'): 

    synonyms_prompt.add(word.lemmas()[0].name())   

synonyms_prompt
```

```python
# sentence with plagiarism has been identified.

sent = 'The data generated by it was good and up to the mark.'
```

```python
synonyms_good = set()

for word in wordnet.synsets('good'): 

    synonyms_good.add(word.lemmas()[0].name())   

synonyms_good
```

```python
# How to replace any word from the sentence with it's synonym?

for word in list(synonyms_good):
    
    print(sent.replace('good', word))
```

#### Read a word from user and print all the meaning of it, except it's own name

```python
word_sys = input("Enter a word to find the meaning : ")

for synset in wordnet.synsets(word_sys):
    
    if word_sys not in synset.name():
        
        print(synset.name(),synset.lemmas()[0].name(), synset.definition())
        
        print()
```

####  Read a word from user and print all noun meanings from it.

```python
word_sys = input("Enter a word to find the meaning : ")

for synset in wordnet.synsets(word_sys):
    
    if '.n.' in synset.name():
        
        print(synset.name(),synset.lemmas()[0].name(), synset.definition())
        
        print()
```

#  Read a word from user and print all adjective meanings from it.

```python
sent=''' The fibre is most often spun into yarn or thread and used 
to make a soft,rigid and flexible textile.  '''
```

```python
token=word_tokenize(sent)
token
```

```python
adj=[word[0] for word in pos_tag(token) if word[1].startswith('JJ')]
adj
```

```python
for word in adj:
    for synset in wordnet.synsets(word):

        
            print(synset.name(),synset.lemmas()[0].name())
        
            print()
```

# Antonyms or opposite word

```python
ant=set()
for synset in wordnet.synsets("run"):
    for lemma in synset.lemmas():
        if lemma.antonyms():
            ant.add(lemma.antonyms()[0].name())
print('Antonymes',ant)
```
