## Transfer Learning

### Imports

```python
### -------------------
### Importing libraries
### -------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Flatten,Conv2D,MaxPool2D,Dropout
from keras.preprocessing.image import ImageDataGenerator
```

### Global variables

### Create the image generator object to read images

```python
train_datagen= ImageDataGenerator(rescale=1/255,
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1/255)
```

# Read the images

```python
train_dataset = train_datagen.flow_from_directory('/home/dai/7. NLP & CV/NLP/Imagenet_new/train', 
                                         class_mode = 'categorical',
                                         target_size = (64,64),
                                          batch_size=32
                                         )
```

```python
train_dataset.class_indices
```

```python
train_dataset.filenames
```

# Model creation

```python
def build(input_shape=(64,64,3),num_classes=10):
    """Defines a CNN model with a specified
    input shape and number of classes."""
    model = Sequential()

    model.add(Conv2D(32,(3,3),
                input_shape =input_shape, activation = 'relu')
         )

    model.add(MaxPool2D((2,2)))
    
    model.add(Conv2D(64,(3,3), activation = 'relu'))

    model.add(MaxPool2D((2,2)))
    
    model.add(Flatten())
    model.add(Dense(num_classes, activation = 'softmax'))
    return model
```

```python
model=build()
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_dtype=True,show_layer_activations=True,show_layer_names=True,dpi=200)
```

# Compile model

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

# Train the model

```python
model.fit(train_dataset, epochs = 10,
         batch_size = 10)
```

# evaluate the model

```python
model.evaluate(train_dataset)
```
