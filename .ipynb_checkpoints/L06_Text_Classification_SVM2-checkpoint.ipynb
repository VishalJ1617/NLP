# Data Prepration and cleanung

```python
sent="Hello friends!,how are you?,Welcome to python Programming"
```

```python
tokens1=word_tokenize(sent)
```

```python
tokens1
```

```python
stop=stopwords.words('english')
stop
```

```python
ls=LancasterStemmer()
```

```python
tokens2=[i.lower() for i in tokens1 if i not in string.punctuation]
tokens3=[ls.stem(i) for i in tokens2 if i not in stop]
tokens3
```

```python
def clean_text(text):
    tokens1=word_tokenize(text)
    tokens2=[i.lower() for i in tokens1 if i.isalpha() or i.isdigit()]
    tokens3=[ls.stem(i) for i in tokens2 if i not in stop]
    return tokens3
```

```python
clean_text(sent)
```

```python
if 'on' in stop:
    print('yes')
```

```python
sent2="When we visited the ground no one was playing."
clean_text(sent2)
```

```python
df['text'].apply(lambda x:clean_text(x))
```

# TF*IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(analyzer=clean_text)
```

```python
x_vect=tfidf.fit_transform(df['text'])
```

```python
x_vect
```

```python
tfidf.get_feature_names_out().size
```

```python
y=df['label']
```

# Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_vect,y,random_state=23,stratify=y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Build the Model

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,classification_report
```

```python
classifier=SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
#Prediction on model
y_pred=classifier.predict(x_test)
y_pred
```

# Performance evaluation

```python
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
