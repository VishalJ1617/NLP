```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

### Global_variables

```python
RANDOM_STATE = 42
```

```python
features, labels = datasets.make_moons(n_samples = 1000, noise = 0.2, random_state = RANDOM_STATE)
```

```python
features.shape
```

```python
labels.shape
```

```python
data_df = pd.DataFrame(features, columns=['X1', 'X2'])
```

```python
data_df['target'] = labels
```

```python
data_df
```

```python
data_df.plot(kind = 'scatter', x = 'X1', y = 'X2', cmap = 'autumn')
plt.show()
```

```python
sns.scatterplot(data=data_df, x='X1', y='X2', hue='target', s=data_df['X1']*10)
plt.show()
```
