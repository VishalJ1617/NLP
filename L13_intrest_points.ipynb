```python
#Importing libraries
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
#Get the image
img=cv2.imread('images/dog.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

# Extract intrest points

```python
#Convert the image into gray scale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#Create the object of SIFT algorithm
sift=cv2.SIFT_create()

#Detect and compute the keypoints
kp,_=sift.detectAndCompute(gray,None)

#Draw keypoints on image and highlight
new_img=cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
len(kp)
```

```python
plt.imshow(new_img)
```
